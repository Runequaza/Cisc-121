```python
class Solution(object):
    def countStudents(self, students, sandwiches):
        from collections import deque
        student = deque(students)
        sandwich = deque(sandwiches)
        counter = len(student)
        while len(student) > 0:
            if student[0] == sandwich[0]:
                student.popleft()
                sandwich.popleft()
                counter = len(student)
            else:
                counter-= 1
                if counter == 0:
                    return len(student)
                student.append(student.popleft())
        return 0
        #https://leetcode.com/problems/number-of-students-unable-to-eat-lunch/solutions/2510921/python-queue-easy-approach-by-imparikalp-qc15/?envType=problem-list-v2&envId=queue Looked at but didnt use
```
